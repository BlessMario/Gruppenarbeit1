```python
import pandas as pd
import openpyxl
import xlrd
```

```python
df_post = pd.read_csv('./Daten/post.csv', sep=';')
```

```python
df_post.head()
```

```python
# Löschen der Column Names Übersetzungen
del_columns =df_post.columns[df_post.columns.str.endswith(('_fr','_it','_en'))]
df_post = df_post.drop(del_columns, axis = 1)
```

```python
df_post.shape
```

```python
#Löschen nicht brauchbarer columns
df_post = df_post.drop(['id',
 'poitypeid',
 'address_countrycode',
 'address_technicalplz6',
 'address_gemeindeid',
 'contact_phone',
 'contact_fax',
 'contact_email_addresse_de',
 'additionaldescription',
 'offerpackage',
 'openinghoursevent',
 'accessinformation',
 'pickuptime_description_de',
 'pickuptime_time',
 'service_servicetypeid',
 'receipttype_de',
 'forprivatecustomer',
 'forbusinesscustomer',
 'url_de',
 'urlmobil_de'], axis=1)
```

```python
df_post.to_excel('post checkup.xlsx')
```

```python
#Split der geopoint daten in 2 separate Columns. Geopoint kann danach gelöscht werden
df_post[['post WGS84_E','post WGS84_N']] = df_post['geopoint'].str.split(pat=',', expand=True)
df_post = df_post.drop(['geopoint'], axis=1)
```

```python
#NaN Werte aus geodaten entfernen -> ganze row wird gelöscht -> betrifft nur 11 Zeilen
df_post = df_post.dropna(subset=['post WGS84_E'])
```
